```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import itertools as it
```

```python
def show_image(image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
image = cv2.imread("../assets/pieces_real_phone/01.jpeg")
show_image(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
show_image(gray_image) # Original grayscale image

# 1. Create a mask: Identify all pixels in the grayscale image that are exactly black (intensity 0).
# The result, black_mask, is a boolean NumPy array.
black_mask = (gray_image == 0)

# 2. Modify the image: Create a modified copy of the grayscale image.
modified_gray = gray_image.copy()

# 3. Apply the mask: Set all pixels that were black in the original (where black_mask is True) to 255 (white).
# This effectively 'masks' out the black pixels from the subsequent THRESH_BINARY_INV.
modified_gray[black_mask] = 255
show_image(modified_gray) # Grayscale image with original black pixels now white

# 4. Apply the inverse threshold on the modified image.
# This will now only highlight pixels with intensity < 200, excluding the original black areas (now 255).
_, thresh = cv2.threshold(modified_gray, 200, 255, cv2.THRESH_BINARY_INV)
show_image(thresh)
```
